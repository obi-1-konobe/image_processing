```python
for image in [img_1, img_2, img_3]:
    top_border = int(image.shape[0] * 0.1)
    bottom_border = int(image.shape[0] * 0.9)
    left_border = int(image.shape[1] * 0.1)
    right_border = int(image.shape[1] * 0.9)
    image = image[top_border: bottom_border, left_border: right_border]
```
