```python
%matplotlib inline

from skimage.io import imread, imsave, imshow
from skimage import img_as_float
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = imread('images/tiger-low-contrast.png')
imshow(img)
```

```python
img_array = np.ravel(img)
```

```python
plt.hist(img_array, range(257))
plt.show
```

```python
contrast_img = (img - img.min()) * 255. / (img.max() - img.min())
```

```python
contrast_img.max()
```

```python
contrast_img = contrast_img.astype('uint8')
```

```python
imshow(contrast_img)
```

```python
contrast_array = np.ravel(contrast_img)
```

```python
plt.hist(contrast_array, range(257))
plt.show
```

```python
imshow(img)
```

```python
n_pix = img.shape[0] * img.shape[1]
n_pix
```

```python
k = round(0.05 * n_pix)
k
```

```python
values, bin_edges, patches = plt.hist(img.ravel(), bins=range(257))
```

```python
values
```

```python
bin_edges
```

```python
counter = 0
for i in range(256):
    counter += values[i]
    if counter > k:
        x_min = i
        break
```

```python
x_min
```

```python
counter = 0
for i in range(255, -1, -1):
    counter += values[i]
    if counter > k:
        x_max = i
        break
```

```python
x_max
```

```python
stable_contrast = (img - x_min) * 255. / (x_max - x_min)
```

```python
stable_contrast
```

```python
stable_contrast = stable_contrast.astype('uint8')
```

```python
stable_contrast.max()
```

```python
imshow(stable_contrast)
```
