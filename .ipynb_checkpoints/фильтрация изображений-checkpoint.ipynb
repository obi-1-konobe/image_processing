```python
from itertools import product
```

```python
some_input = input()
sigma = float(some_input)
```

```python
k = round(3 * sigma * 2 + 1)
k
```

```python
idx = int((k - 1) / 2)
coordinat_y_x = list(product(range(idx, -idx - 1, -1), range(-idx, idx + 1)))
coordinat_y_x
```

```python
kernel = np.full((k, k), -np.inf)
kernel
```

```python
def gausian(sigam, x, y):
    return np.exp(-(x ** 2 + y ** 2) / (2 * (sigma ** 2))) / (2 * np.pi * (sigma ** 2))
```

```python
for i in range(k):
    part = coordinat_y_x[i * k: i * k + k]
    row = []
    for coordinat in part:
        y, x = coordinat
        result = gausian(sigma, x, y)
        row.append(result)
    kernel[i] = row
    
```

```python
kernel = np.round(kernel / np.sum(kernel), 5)
kernel
```

```python
for i in kernel:
    print(*i)
    
```

```python
from skimage.io import imread, imshow, imsave
from scipy.signal import convolve2d
```

```python
img = imread('images/tiger-gray-small.png')
imshow(img)
```

```python
sigma = 0.66
k = round(sigma * 3 * 2 + 1)
k
```

```python
kernel_list = []
for y in range(k // 2, -k // 2, -1):
    for x in range(-k // 2 + 1, k // 2 + 1):
        kernel_list.append(gausian(sigma, x, y))
```

```python
kernel = np.array(kernel_list).reshape((k, k))
```

```python
kernel /= np.sum(kernel)
kernel
```

```python
print(np.round(kernel, 4))
```

```python
new_img = convolve2d(img, kernel, mode='valid')
```

```python
new_img = new_img.astype('uint8')
```

```python
imshow(new_img)
```

```python
true_img = imread('images/gaussian-tiger.png')
imshow(true_img)
```

```python
np.array_equal(true_img, new_img)
```

```python
new_kernel = np.zeros_like(img)
```

```python
new_kernel = new_kernel.astype('float')
```

```python
new_kernel[:k, :k] = kernel
```

```python
new_kernel
```

```python
y, x = img.shape
```

```python
result_list = []
img = img.astype('float')
for i in range(0, -(y - k + 1), -1):
    for j in range(0, -(x - k + 1), -1):
        new_mtx = np.roll(img, i, 0)
        new_mtx = np.roll(new_mtx, j, 0)
        result = np.sum(new_mtx * new_kernel)
        result_list.append(result)
```

```python
new_img_2 = np.array(result_list).reshape((296, 396))
```
