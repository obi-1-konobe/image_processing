```python
from skimage.io import imread, imshow, imsave
from skimage import img_as_float
img = imread('images/tiger-color.png')
```

```python
img_f = img_as_float(img)
```

```python
red = img_f[:, :, 0]
```

```python
imshow(red)
```

```python
green = img_f[:, :, 1]
imshow(green)
```

```python
blue = img_f[:, :, 2]
imshow(blue)
```

```python
from numpy import dstack
```

```python
img_combine = dstack((red, green, blue))
```

```python
imshow(img_combine)
```

```python
avg_gray = (red + green + blue) / 3
imshow(avg_gray)
```
