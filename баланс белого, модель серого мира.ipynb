```python
%matplotlib inline

from skimage.io import imread, imshow, imsave
from skimage import img_as_float, img_as_ubyte
import numpy as np
```

```python
img = imread('images/railroad.png')
imshow(img)
```

```python
img_f = img_as_float(img)
```

```python
red = img_f[:, :, 0]
green = img_f[:, :, 1]
blue = img_f[:, :, 2]
```

```python
avg = (np.average(red) + np.average(blue) + np.average(green)) / 3
avg
```

```python
r_w = np.average(red) / avg
g_w = np.average(green) / avg
b_w = np.average(blue) / avg
r_w, g_w, b_w
```

```python
new_red = red / r_w
new_green = green / g_w
new_blue = blue / b_w
```

```python
new_img = np.dstack((new_red, new_green, new_blue))
new_img = np.clip(new_img, 0, 1)
new_img = img_as_ubyte(new_img)
```

```python
imshow(new_img)
```

```python
imsave('images/railroad_gray_world.png', new_img)
```
