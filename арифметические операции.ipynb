```python
from skimage.io import imread, imshow, imsave
```

```python
img = imread('images/tiger.jpg')
```

```python
img.dtype
```

```python
from skimage import img_as_float
```

```python
img_f = img_as_float(img)
```

```python
img.min(), img.max()
```

```python
img_f.min(), img_f.max()
```

```python
imshow(img_f / 1.5)
```

```python
imshow(img_f / 4)
```

```python
imshow(img_f / 4 + 0.25)
```

```python
from numpy import clip
```

```python
imshow(clip(img_f * 1.5, 0, 1))
```

```python
imshow(clip(img_f * 2 - 0.5, 0, 1))
```
