```python
%matplotlib inline

from skimage.io import imread, imshow, imsave
import numpy as np
```

```python
img = imread('images/tiger-gray-small.png')
imshow(img)
```

```python
k = 7
y, x = img.shape
```

```python
result_list = []
for i in range(y - k + 1):
    for j in range(x - k + 1):
        kernel = img[i: i + k, j: j + k]
        median = np.median(kernel.ravel())
        result_list.append(median)
```

```python
new_img = np.array(result_list).reshape((y - k + 1, x - k + 1))
```

```python
new_img = np.clip(new_img, 0, 255).astype('uint8')
```

```python
imshow(new_img)
```
